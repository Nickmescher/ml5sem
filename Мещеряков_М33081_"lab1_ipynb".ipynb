Напишите программу, которая будет решать систему линейных уравнений вида:

\begin{matrix} a_{11}x+a_{12}y=b_{1}&(1) \\ a_{21}x+a_{22}y=b_{2} &(2) \end{matrix}
​	
На вход программа должна принимать 2 строки.

Каждая строка содержит 3 действительных числа через пробел: коэффициенты a и b.

На выходе программа должна выдавать значения x и y в одну строку через пробел.
 

Примечание 1. Для разделения строки на отдельные элементы используйте функцию split()

Примечание 2. В качестве разделителя между целой и дробной частью используйте точку, например 0.2

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def linan(A, B):
  if A[0][0]/A[0][1] != A[1][0]/A[1][1]:
    answ = np.linalg.solve(A, B)
    return answ
  elif A[0][0]/A[1][0] == A[0][1]/A[1][1] and A[0][0]/A[1][0] == B[0]/B[1]:
    return "Бесконечное количество решений"
  else:
    return "Нет решений"

eq1 = input()
eq2 = input()
eq1 = eq1.split()
eq2 = eq2.split()
Amat = np.array([[eq1[0], eq1[1]],[eq2[0], eq2[1]]]).astype(np.float32)
Bmat = np.array([eq1[2], eq2[2]]).astype(np.float32)
def f1(x):
  return (Amat[0][0]*x)/Amat[0][1] + Bmat[0]/Amat[0][1]

def f2(x):
  return (Amat[1][0]*x)/Amat[1][1] + Bmat[1]/Amat[1][1]
x1 = -10
x2 = 20
plt.plot([x1, x2], [f1(x1), f1(x2)], label="f1")
plt.plot([x1, x2], [f2(x1), f2(x2)], label="f2")
plt.xlabel('X') 
plt.ylabel('Y') 
print(linan(Amat, Bmat))
````

# Задание 3

1) реализуйте 2 функции:

$f(x) = (x+a)^2 - b$ 

$g(x) = |f(x)|$

2) Найдите аргументы, при которых достигается минимум функций $f(x)$ и $g(x)$

Пример. 
Входные значение: $a=3$ и $b=4$

Выходные значения:

`-3.0 -1.0`

````python
inp1 = int(input())
inp2 = int(input())

def f1(a, b):
  answ1y = []
  answ1x = []
  for x in range(-10, 10):
    answ1y.append((x+a)**2 - b)
    npansw1 = np.array(answ1y)
    answ1x.append(x)
    npansw1x = np.array(answ1x)
  return npansw1, npansw1x

def f2(f1, a , b):
  answ2 = f1(a,b)
  answ2y = np.absolute(answ2[0])
  answ2x = answ2[1]
  return answ2y, answ2x

res1 = f1(inp1,inp2)
res2 = f2(f1,inp1,inp2)
ymin1 = np.amin(res1[0])
answer1 = []
for n in range(len(res1[0])):
  if res1[0][n] == ymin1:
    answer1.append(res1[1][n])

ymin2 = np.amin(res2[0])
answer2 = []
for n in range(len(res2[0])):
  if res2[0][n] == ymin2:
    answer2.append(res2[1][n])

print(answer1, answer2)
````

# Задание 5

1000 человек бросают монетку. Каждый бросил 10 раз подряд.

 1. Какова вероятность того, что хотя бы у одного человека монетка выпала 10 раз подряд той же самой стороной (орлом или решкой)? 
 2. Какова вероятность того, что хотя бы у одного человека монетка выпала 8 раз подряд той же самой стороной (орлом или решкой)? 
 3. Какова вероятность того, что ровно у троих человек из 1000 монетка выпала 10 раз подряд той же самой стороной (орлом или решкой)?

Снимок экрана 2022-09-20 025143.png

````python
#task1

p = 2 / 2**10
q = 1 - p
ans1 = 0
people = 1000

def combination(n,k):
  return np.math.factorial(n) / (np.math.factorial(n-k) * np.math.factorial(k))

for k in range(1, 1001):
  ans1 += combination(people, k) * p**k * q**(people-k)

print(ans1)

#task2

p = 2*5 / 2**10
q = 1 - p
ans2 = 0

for k in range(1, 1001):
  ans2 += combination(people, k) * p**k * q**(people-k)

print(ans2)

#task3
p = 2 / 2**10
q = 1 - p
ans3 = combination(people, 3) * p**3 * q**(people-3)

print(ans3)
````
